```python
x=np.arange(-5.0,5.0,0.01)
f=(0.5*(x**4-16.0*x**2+5.0*x)+39.16599)/39.16599
#f=8.347*(x**4-16.0*x**2+5.0*x+2.0*39.16599)/(2.0*39.16599)

# Plot.     
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 30})

plt.subplots(1, figsize=(12,12))
plt.plot(x,np.exp(-f),'k')
plt.grid()
plt.savefig('styblinski.png',format='png')
plt.show()

dx=x[1]-x[0]
np.sum(np.exp(-f))*dx
```
