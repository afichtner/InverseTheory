```python
import numpy as np
import matplotlib.pyplot as plt
import testfunctions
```

```python
# Model space dimension.
dim=1000

# Dimensions for plotting.
dim1=0
dim2=999

# Type of test function.
test_function='rastrigin_2'

# Value range.
m1_min=-4.0
m1_max=4.0
m2_min=-32.0
m2_max=32.0
```

```python
# Initialise function and arrays.

fct=testfunctions.f(dim,test_function)

m=np.zeros(dim)

m1=np.linspace(m1_min,m1_max,300)
m2=np.linspace(m2_min,m2_max,300)

m1v,m2v=np.meshgrid(m1,m2)
U=np.zeros(np.shape(m1v))
```

```python
fct.A.toarray()
```

```python
# March through grid to evaluate function.

for i in range(len(m1)):
    for j in range(len(m2)):
        
        m[dim1]=m1[i]
        m[dim2]=m2[j]
        U[j,i]=fct.U(m)

# Subtract minimum for better plotting.
U-=np.min(np.abs(U))
        
# Plot.     
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 30})
plt.subplots(1, figsize=(12,12))

plt.pcolor(m1v,m2v,np.exp(-U),cmap='Greys')
#plt.pcolor(m1v,m2v,U,cmap='Greys')
plt.colorbar()
plt.contour(m1v,m2v,np.exp(-U),10,colors='k')
plt.xlabel('m %d' % dim1)
plt.ylabel('m %d' % dim2)
plt.grid()
plt.savefig('testfunction.png',format='png')
plt.show()
```
