```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(-5.0,5.0,0.01)
f=10.0+x**2-10.0*np.cos(2.0*np.pi*x)

# Plot.     
plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 30})

plt.subplots(1, figsize=(12,12))
plt.plot(x,np.exp(-f),'k')
plt.grid()
plt.savefig('rastrigin_pdf.png',format='png')
plt.show()
```

```python
df=2.0*x**3-16.0*x+5

plt.plot(x,df)
plt.show()

ddf=6.0*x**2-16

plt.plot(x,ddf)
plt.show()
```
