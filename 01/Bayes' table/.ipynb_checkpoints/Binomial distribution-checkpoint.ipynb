# Binomial distribution

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.special as special

plt.rcParams["font.family"] = "serif"
plt.rcParams.update({'font.size': 28})
```

```python
n=10
```

```python
p=np.arange(0,n+1,1)


alpha=0.1
P1=np.zeros(len(p))
for i in p:
    P1[i]=float(np.math.factorial(n)/(np.math.factorial(n-i)*np.math.factorial(i)))*np.power(alpha,i)*np.power(1.0-alpha,n-i)

alpha=0.5
P2=np.zeros(len(p))
for i in p:
    P2[i]=float(np.math.factorial(n)/(np.math.factorial(n-i)*np.math.factorial(i)))*np.power(alpha,i)*np.power(1.0-alpha,n-i)

alpha=0.8
P3=np.zeros(len(p))
for i in p:
    P3[i]=float(np.math.factorial(n)/(np.math.factorial(n-i)*np.math.factorial(i)))*np.power(alpha,i)*np.power(1.0-alpha,n-i)


    
    
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(30,10))

ax1.plot(p,P1,'ko',markersize=7)
ax1.plot(p,P1,'k--',linewidth=1.5)
ax1.set_xlim([0,n])
ax1.grid()
ax1.set_xlabel('p')
ax1.set_ylabel('P(p)',labelpad=20)
ax1.set_title(r'$\alpha=0.1$',pad=15)

ax2.plot(p,P2,'ko',markersize=7)
ax2.plot(p,P2,'k--',linewidth=1.5)
ax2.set_xlim([0,n])
ax2.grid()
ax2.set_xlabel('p')
ax2.set_title(r'$\alpha=0.5$',pad=15)

ax3.plot(p,P3,'ko',markersize=7)
ax3.plot(p,P3,'k--',linewidth=1.5)
ax3.set_xlim([0,n])
ax3.grid()
ax3.set_xlabel('p')
ax3.set_title(r'$\alpha=0.8$',pad=15)


plt.savefig("binomial.pdf", bbox_inches='tight')
plt.show()
```
